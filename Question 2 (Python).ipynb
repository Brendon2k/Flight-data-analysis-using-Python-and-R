Importing packages

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

Importing the main dataset

```python
#Using the pandas read_csv() function to import the dataset
full_dataset = pd.read_csv("../datasets/dataset_full_new.csv")
full_dataset
```

Removing the unnamed column as it adds no value to our analysis

```python
# To remove the unnamed column we use the iloc function to select all rows and columns except the first column (Unnamed column)
full_dataset = full_dataset.iloc[:, 1:]
full_dataset
```

Changing Jupyter notebook settings to view all columns

```python
# changing settings to display all columns
pd.set_option("display.max_columns", None)
# displaying the main dataset head
full_dataset.head()
```

Importing the plane dataset

```python
#Using the pandas read_csv() function to import the dataset
plane_dataset =pd.read_csv("../datasets/plane-data.csv")
plane_dataset
```

Checking for null values in the plane dataset

```python
# Using the pandas isnull() function to find the count of null values present in each column
pd.isnull(plane_dataset).sum(axis=0)
```

Identifying the column names in the full_dataset

```python
full_dataset.columns
```

Identifying the column names in the plane_dataset

```python
plane_dataset.columns
```

Removing unnecessary columns in the full_dataset

```python
full_dataset = full_dataset.filter(['TailNum', 'ArrDelay', 'DepDelay'], axis=1)
full_dataset
```

Removing unnecessary columns in the plane_dataset

```python
plane_dataset = plane_dataset.filter(['tailnum', 'year'])
plane_dataset
```

Checking for null values in the filtered plane_dataset

```python
# Using the pandas isnull() function to find the count of null values present in each column
pd.isnull(plane_dataset).sum(axis=0)
```

Removing the null values in the filtered plane_dataset

```python
# Using the pandas notnull() function to only select non-null values from 'year' column
plane_dataset = plane_dataset[pd.notnull(plane_dataset['year'])]
plane_dataset
```

```python
# pd.merge(left, right, on=['key1', 'key2'],how='left')
```

Making a copy of plane_dataset to avoid "A value is trying to be set on a copy of a slice from a DataFrame" error

```python
# Making a copy of the entire plane_dataset dataframe
plane_dataset_1 = plane_dataset.iloc[:,:].copy()
```

Renaming the plane_dataset_1 tail number column to match that of the full_dataset tail number column so that merging is possible. 

```python
#renaming the tailnum column in plane_dataset_1 so that merging is possible
plane_dataset_1.rename(columns={"tailnum":"TailNum"}, inplace = True)
print(plane_dataset_1)
```

Merging plane_dataset_1 with the full_dataset based on Tail Number

```python
# Using pandas merge() function to preform an inner join merge based on TailNum
merged_dataset = pd.merge(full_dataset, plane_dataset_1, on=['TailNum'])
merged_dataset
```

Checking for null values in the merged_dataset

```python
pd.isnull(merged_dataset).sum()
```

Checking the dtypes of the merged_dataset

```python
merged_dataset.dtypes
```

Since the year is an object we check the year column for non-numerical values and unfeasible year values

```python
#Checking the value counts for the year column in the merged_dataset
merged_dataset['year'].value_counts()

# We found that there are years named 'None' and '0000'
```

Removing the '0000' and 'None' values from the year column in the merged_dataset

```python
# Using the pandas drop() function to drop a row if it either contains '0000' or 'None' in the year column of the merged_dataset
merged_dataset.drop(merged_dataset[(merged_dataset['year'] == '0000') | (merged_dataset['year'] == 'None') ].index, inplace = True)
merged_dataset
```

Confirming whether '0000' and 'None' year value containing rows were removed

```python
merged_dataset['year'].value_counts()
```

Converting the year variable from an object to an integer

```python
# Converting the year column in the merged_dataset to an integer using the astype() function
merged_dataset['year'] = merged_dataset['year'].astype(int)
```

Confirming whether year was converted to integers

```python
merged_dataset.dtypes
```

Constructing a line plot to compare the average arrival delay against plane manufacture year

```python
# Line plot for average arrival delays against plane manufacture year
fig, ax = plt.subplots(1, 1, figsize = (20, 8))

# Making a variable which holds the dataframe containing the mean Arrival Delay (ArrDelay) grouped by Plane Manufactured Year (year)
avg_ArrDelay_plane_year = merged_dataset[['year','ArrDelay']].groupby(['year']).mean()

# Plotting the line graph
ax.plot(avg_ArrDelay_plane_year,
        color='#155ff0',
        alpha = 0.75)

# Setting the title
ax.set_title("Line Plot for Average Arrival Delay by Plane Manufactured Year",
                color='#2c2f33',
                family = 'monospace',
                size = 25,
                weight = 'bold')

# Creating the x-axis labels
ax.set_xlabel('Plane Manufactured Year',
                 color='#2c2f33',
                 family = 'monospace',
                 size = 18,
                 labelpad=7)

# Creating the y-axis labels
ax.set_ylabel('Arrival delay (Time in minutes)',
                 color='#2c2f33',
                 family = 'monospace',
                 size = 18,
                 labelpad=10)

# Formatting the x and y axis tick labels
ax.tick_params(axis='x', labelcolor='#ff6361', labelsize=15)
ax.tick_params(axis='y', labelcolor='#ffa600', labelsize=15)

# Creating x axis ticks starting from 1955 to 2010 with 5 year intervals
plt.xticks(np.arange(1955, 2011, 5))

# Adjusting the subplot layout parameters
plt.subplots_adjust(left=0.15)

# Creating the overall average arrival delay variable
avgDelay = np.sum(merged_dataset['ArrDelay']) / len(merged_dataset['ArrDelay'])
# Plotting the overall average arrival delay line so that there is a baseline for making comparisons
plt.axhline(y = avgDelay, color='#f44336', label = 'Overall average delay', linestyle = 'dashed')

# Showing legend
plt.legend()

# Saving the graph
plt.savefig("Line plot for average arrival delay by plane manufacture year.png")
plt.show()
```
