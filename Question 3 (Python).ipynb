## Data wrangling

Importing packages

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

Changing Jupyter notebook settings to view all columns

```python
pd.set_option("display.max_columns", None)
```

Importing the main dataset

```python
#Using the pandas read_csv() function to import the dataset
full_dataset = pd.read_csv("../datasets/dataset_full_new.csv")
full_dataset
```

Removing the unnamed column as it adds no value to our analysis

```python
# To remove the unnamed column we use the iloc function to select all rows and columns except the first column (Unnamed column)
full_dataset = full_dataset.iloc[:, 1:]
full_dataset
```

Importing the airports dataset

```python
#importing the airports dataset
airports = pd.read_csv("../datasets/airports.csv")
airports
```

Checking for null values in airports dataset

```python
# Using the pandas isnull() function to find the count of null values present in each column
pd.isnull(airports).sum(axis=0)
```

Finding list of entries in airports dataset where state has null values

```python
airports[pd.isnull(airports['state'])]
```

Checking for duplicate entries in iata column in the airports dataset so that we can cross-reference iata codes and fill in missing values if available

```python
# Using a boolean expression to see if there are duplicates in the iata column in the airports dataset.
boolean = airports['iata'].duplicated().any()
boolean

# No iata duplicates are found. Hence cannot cross reference within airports dataset and find city or state of null values.
# Hence we can remove null values in city and state
```

Since there are no iata duplicates to cross-reference and fill in missing data and since all 12 null values are in the state and city column (which are our main areas of interest for this question) we can go ahead and remove all null values now itself.

```python
#removing null values in the airports dataset
airports = airports[pd.notnull(airports["state"])]
airports
```

```python
# Checking if all null values were removed
pd.isnull(airports).sum(axis=0)
```

Checking how many states and cities are present in the airports dataset

```python
# Finding number of cities
airports['city'].value_counts()
```

```python
# Finding number of states
airports['state'].value_counts()

# There are 56 states and 2675 cities
# Therefore states was selected to compare locations
```

Dropping unnecessary columns in the full_dataset

```python
# Using the filter() function to only keep the necessary columns (Year, Origin and Dest)
full_dataset = full_dataset.filter(['Year', 'Origin', 'Dest'], axis = 1)
full_dataset
```

Dropping unnecessary columns in the airports dataset

```python
# Using the filter() function to only keep the necessary columns (iata and state)
airports_1 = airports.filter(['iata', 'state'], axis = 1)
airports_1
```

We have now filtered the airports dataset only selecting the variables that we need for the merge. In order to find BOTH Origin state and Destination state of the airports we will need a copy of the airports_1 dataframe

```python
# Making a copy of the airports_1 dataframe
airports_2 = airports_1.iloc[:,:].copy(deep = True)
airports_2
```

We use the airports_1 dataframe to find the Origin State and the airports_2 dataframe to find the Destination State of each flight

Now we rename the columns of the airports_1 dataframe so that they match the full_dataset.
('iata' becomes 'Origin' and 'state' becomes 'Origin_State')

```python
# Renaming airports_1 dataframe columns using the rename() function.
# This line of code changes 'iata' to 'Origin' and 'state' to 'Origin_State'
airports_1.rename(columns= {'iata':'Origin', 'state':'Origin_State'}, inplace = True)
airports_1
```

Now we rename the columns of the airports_2 dataframe so that they match the full_dataset. ('iata' becomes 'Dest' and 'state' becomes 'Dest_State')

```python
# Renaming airports_1 dataframe columns using the rename() function
# This line of code changes 'iata' to 'Dest' and 'state' to 'Dest_State'
airports_2.rename(columns= {'iata':'Dest', 'state':'Dest_State'}, inplace = True)
airports_2
```

We now merge full_dataset and airports_1 in order to find the Origin State of each flight in the full_dataset

```python
# Using the pandas merge() function to merge full_dataset and airports_1 using an inner join on the 'Origin'
merged_dataset = pd.merge(full_dataset, airports_1, on = 'Origin')
merged_dataset
```

We now merge merged_dataset and airports_2 in order to find the Destination State of each flight in the merged_dataset

```python
# Using the pandas merge() function to merge merged_dataset and airports_1 using an inner join on the 'Dest'
merged_dataset = pd.merge(merged_dataset, airports_2, on = 'Dest')
merged_dataset
```

Checking to make sure there are no null values in the merged_dataset

```python
# checking for null values in merged_dataset
pd.isnull(merged_dataset).sum(axis=0)
```

Checking if dtype of Year is integer

```python
merged_dataset.dtypes
```

We now split the merged_dataset into two based on year

```python
# Subsetting the merged_dataset values from 2006 into a new dataframe called merged_dataset_2006 using the isin() function
merged_dataset_2006 = merged_dataset[merged_dataset['Year'].isin([2006])]

# Subsetting the merged_dataset values from 2007 into a new dataframe called merged_dataset_2007 using the isin() function
merged_dataset_2007 = merged_dataset[merged_dataset['Year'].isin([2007])]
```

We now find the count of flights that occurred between each state during 2006

In order to do that we group the year-wise subsetted dataframe values by Origin State, obtain their value counts for each Destinaion State, unstack it and fill NA values with zero.

```python
'''
Making the grouped_dataset_2006 dataframe by first grouping merged_dataset_2006 by 'Origin_State', then obtaining the count of 
flights between each destination state using the value_counts() function and then unstacking to reshape the dataframe into 
an unstacked form and finally filling the NA values with zero
'''
grouped_dataset_2006 = merged_dataset_2006.groupby('Origin_State').Dest_State.value_counts().unstack().fillna(0)
grouped_dataset_2006
```

```python
'''
Making the grouped_dataset_2007 dataframe by first grouping merged_dataset_2007 by 'Origin_State', then obtaining the count of 
flights between each destination state using the value_counts() function and then unstacking to reshape the dataframe into 
an unstacked form and finally filling the NA values with zero
'''
grouped_dataset_2007 = merged_dataset_2007.groupby('Origin_State').Dest_State.value_counts().unstack().fillna(0)
grouped_dataset_2007
```

Checking the shape of the year-wise grouped dataframes

```python
grouped_dataset_2007.shape
```

```python
grouped_dataset_2006.shape
```

Since the year-wise grouped dataframes have similar shapes, we can subtract both of them to find the change in number of flights that occured between 2006 and 2007

```python
# Creating a dataframe called subtracted_dataset which captures the change in number of flights
# Subtracting each cell from the grouped_dataset_2007 dataframe by each corresponding cell in the grouped_dataset_2006 dataframe
subtracted_dataset = grouped_dataset_2007.iloc[:] - grouped_dataset_2006.iloc[:]
subtracted_dataset
```

Saving the subtracted_dataset

```python
# The subtracted_dataset is saved as a csv file using the pandas to_csv() function
subtracted_dataset.to_csv("../output/subtracted_dataset_python.csv")
```

## Plotting the heatmap

Now we Visualize the difference in a heatmap

```python
# Heatmap for the change in number of flights between each state between the years 2006 and 2007
f, ax = plt.subplots(figsize=(24, 18))

# Plotting the heatmap
sns.heatmap(subtracted_dataset, annot=False, square=False, cmap='RdBu', linewidth=1)

# Setting the title
ax.set_title("Heatmap for change in number of flights between states during the period of 2006 and 2007",
                color='#2c2f33',
                family = 'monospace',
                size = 25,
                weight = 'bold')
plt.show()
```

The above heatmap only visualizes 3 significant areas which all have 4-figure values. Let us introduce a weighted index so that we get more hotspots of other areas with significant changes in number of flights.

Find the largest value for change in number of flights (can be positive or negative)

```python
# Finding the maximum value of subtracted_dataset
subtracted_dataset.max()
```

```python
# Finding the minimum value of subtracted_dataset
subtracted_dataset.min()
```

We can see that Texas (TX) shows the greatest change in number of flights with -20197 flights. Hence we divide each value in the subtracted_dataset dataframe by 20197 to get weighted ratios. Thereby allowing us to restrict our heatmap range from -1 to +1

```python
# Dividing each value in the subtracted_dataset dataframe by 20197
subtracted_dataset_ratios = (subtracted_dataset.iloc[:])/20197
subtracted_dataset_ratios
```

```python
# Heatmap for the change in number of flights between each state between the years 2006 and 2007 represented as weighted ratios
f, ax = plt.subplots(figsize=(24, 18))

# Plotting the heatmap
'''
Here we cap the maximum and minimum threshold at +0.25 and -0.25 respectively so that the locations with a number of change in 
flights that are 25% or higher than that of Texas show up as hotspots on the heatmap 
'''
sns.heatmap(subtracted_dataset_ratios, annot=False, square=False, cmap='RdBu', linewidth=1, vmax=0.25, vmin=-0.25)

# Setting the title
ax.set_title("Heatmap for change in number of flights between states during the period of 2006 and 2007 in RATIOS",
                color='#2c2f33',
                family = 'monospace',
                size = 25,
                weight = 'bold')
# Saving the graph
plt.savefig("Heatmap for change in number of flights between 2006 and 2007 across different locations.png")
plt.show()
```
