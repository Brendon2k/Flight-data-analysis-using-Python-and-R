Checking for errors in  Year, Month, DayofMonth and CRSDepTime columns.

We now combine the Year, Month, DayofMonth and CRSDepTime to form a column which shows the exact date and time of estimated departure of each flight saved in datetime64[ns] format.

We first extract and convert the minutes portion of CRSDepTime into a string

```python
# Selecting the integers in the ones and tens place of CRSDepTime and converting it to a string
Minutes_CRSDepTime = full_dataset['CRSDepTime'].astype(str).str[-2:]
Minutes_CRSDepTime
```

We similarly extract and convert the hours portion of CRSDepTime into a string

```python
# Selecting the integers in the hundreds and thousands place of CRSDepTime and converting it to a string
Hours_CRSDepTime = full_dataset['CRSDepTime'].astype(str).str[:-2]
Hours_CRSDepTime
```

There might be CRSDepTimes where the thousands and hundreds place does not contain a value (for example 45 which indicates 00:45AM). Therefore these will be saved as empty values in the Hours_CRSDepTime variable. Hence we must locate and replace them with zero to maintain consistency and avoid errors when converting to datetime64[ns] format.

```python
#Checking for empty string values in Hours_CRSDepTime
Hours_CRSDepTime[Hours_CRSDepTime == ""] 
```

```python
#Replacing empty values in Hours_CRSDepTime with '0'
Hours_CRSDepTime[Hours_CRSDepTime == ""] = "0"
```

Converting the DayofMonth to a string and zero padding the single digit days so that it can be converted to datetime format using the to_datetime() function.

```python
# Converting DayofMonth to a string using the astype() function
Days = full_dataset.DayofMonth.astype(str)
'''
Zero padding the single digit days using the apply and lambda function to convert it to a properly formatted string so that it
can be passed through the pandas to_datetime() function.
The lambda function iterates through each row in the Days variable and applys the string formatter. The regular expression '{0:0>2}' 
used here indicates that if a string has less than 2 characters then the function should add a 0 infront of it. Thereby zero
padding the Days.
'''
  
Days = Days.apply(lambda x: '{0:0>2}'.format(x))  

# Checking the Days value counts
Days.value_counts()
```

```python
full_dataset["CRSDepTime_Datetime"] = pd.to_datetime(Days.astype(str) + "-" + full_dataset.Month.astype(str) + "-" + full_dataset.Year.astype(str) + "," + Hours_CRSDepTime + ":" + Minutes_CRSDepTime, 
                                                     format='%d-%m-%Y,%H:%M')
full_dataset["CRSDepTime_Datetime"]
```

Creating a TotalDelay Column in the full_dataset dataframe

```python
# Adding the ArrDelay and DepDelay for each row in the full_dataset
full_dataset["TotalDelay"] = full_dataset['ArrDelay'] + full_dataset['DepDelay']
full_dataset
```

We now sort the full_dataset dataframe by time so that all the flights made by each plane are arranged in an organized timeline.

```python
# Sorting the flights by DateTime_CRSDepTime using the sort_values() function
full_dataset = full_dataset.sort_values(by=['CRSDepTime_Datetime'])
full_dataset
```

Creating a PreviousDelay Column in the full_dataset dataframe

```python
 '''
 We create the PreviousDelay column by first grouping the TotalDelay by each TailNum and then shifting the entire column down by
 one row. Hence each row will show the current delay (TotalDelay) and the previous delay (PreviousDelay) of a specific airplane. 
 '''
full_dataset['PreviousDelay'] = full_dataset.groupby('TailNum')['TotalDelay'].shift(periods=1)

# Checking the first 20 values of the full_dataset dataframe
full_dataset.head(20)
```

```python
# Checking if the PreviousDelay Column has been correctly lagged for a random tail number
full_dataset[full_dataset['TailNum'] == 'N657AW'].head(20)
```

Null values can be caused due to two reasons. Firstly due to the lag in the PreviousDelay column each top entry for each tail number will have a NaN value. Secondly the airplanes who only have one flight recorded will not have any previous delays by default and hence will have a NaN value in their PreviousDelay column. Therefore we need to remove these null values.

```python
# Using the pandas isnull() function to find the count of null values present in each column
pd.isnull(full_dataset).sum(axis=0)
```

Removing the null values in the PreviousDelay column

```python
# Using the pandas notnull() function to remove rows containing null values in the PreviousDelays column
full_dataset = full_dataset[pd.notnull(full_dataset['PreviousDelay'])]

# Checking if all null values have been removed
pd.isnull(full_dataset).sum(axis=0)
```

# Cross Tabulation for Presence of Current Delay and Previous Delay

Making a copy of the full_dataset dataframe to avoid "A value is trying to be set on a copy of a slice from a DataFrame." error

```python
full_dataset_1 = full_dataset.copy()
```

Making a delay status column to show whether a specific type of delay is present or absent

```python
# Making a Current Delay Status column which shows whether TotalDelay is greater than or smaller than zero
# If TotalDelay is greater than zero the numpy where() function assigns the status of 'Present' to the CurrentDelayStatus column
# If TotalDelay is less than zero the numpy where() function assigns the status of 'Absent' to the CurrentDelayStatus column
full_dataset_1['CurrentDelayStatus'] = np.where(full_dataset_1['TotalDelay'] > 0, 'Present', 'Absent')

# Making a Previous Delay Status column which shows whether PreviousDelay is greater than or smaller than zero
# If PreviousDelay is greater than zero the numpy where() function assigns the status of 'Present' to the PreviousDelayStatus column
# If PreviousDelay is less than zero the numpy where() function assigns the status of 'Absent' to the PreviousDelayStatus column
full_dataset_1['PreviousDelayStatus'] = np.where(full_dataset_1['PreviousDelay'] > 0, 'Present', 'Absent')

full_dataset_1
```

Making a cross tabulation of the variables CurrentDelayStatus and PreviousDelayStatus

```python
# Making a cross tabulation describing the status of CurrenDelay and PreviousDelay  using the pandas crosstab() function
full_dataset_1_crosstab = pd.crosstab(full_dataset_1.PreviousDelayStatus, full_dataset_1.CurrentDelayStatus, rownames=["Previous Delay"], colnames=["Current Delay"])
full_dataset_1_crosstab
```

```python
# Getting the summation across rows
devisor = full_dataset_1_crosstab.sum(axis=1)
print(devisor)

# Dividing each value in full_dataset_1_crosstab by it corresponding row total and then multiplying by 100
full_dataset_1_crosstab_percentage = full_dataset_1_crosstab.divide(devisor, axis=0)*100 
full_dataset_1_crosstab_percentage
```
