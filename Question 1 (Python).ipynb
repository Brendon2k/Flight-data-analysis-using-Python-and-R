# Preparing the dataset

Importing the required packages

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

Importing the main dataset

```python
#Using the pandas read_csv() function to import the dataset
full_dataset = pd.read_csv("../datasets/dataset_full_new.csv")
full_dataset
```

Changing Jupyter notebook settings to view all columns

```python
# changing settings to display all columns
pd.set_option("display.max_columns", None)
# displaying the dataframe head
full_dataset.head()
```

Checking available columns

```python
full_dataset.columns
```

Removing the unnamed column as it adds no value to our analysis

```python
# To remove the unnamed column we use the iloc function to select all rows and columns except the first column (Unnamed column)
full_dataset = full_dataset.iloc[:, 1:]
full_dataset
```

Finding the correlation of ArrDelay and DepDelay

```python
#Plotting a pearson Correlation matrix for ArrDelay and DepDelay
correlation_matrix = full_dataset.loc[:, ['ArrDelay', 'DepDelay']].corr('pearson')
correlation_matrix

# It can be seen that ArrDelay and DepDelay are highly correlated
```

# Best time of year to fly to minimize delays

Checking for errors in Month column

```python
#Checking for smallest and largest values aswell as outliers in Months column
print('The smallest value for Months is', full_dataset.Month.min())
print('The largest value for Months is', full_dataset.Month.max())

full_dataset.Month.value_counts()
```

Constructing a dataframe for mean and median Arrival Delay and Departure Delay which are grouped by Months

```python
#Finding table for mean and median Arrival Delay and Departure Delay for each month
full_dataset.groupby('Month').aggregate(mean_ArrDelay = ('ArrDelay', np.mean), 
                                        median_ArrDelay = ('ArrDelay', np.median),
                                        mean_DepDelay = ('DepDelay', np.mean),
                                        median_DepDelay = ('DepDelay', np.median))
```

### Constructing a line plot to show how average arrival delay varies across months

```python
# Line plot for average delays by month
fig, ax = plt.subplots(1, 1, figsize = (20, 8))


# Making a variable which holds the dataframe containing the mean Arrival Delay (ArrDelay) grouped by Month
avg_ArrDelay_Month = full_dataset[['Month','ArrDelay']].groupby(['Month']).mean()

# Plotting the line graph
ax.plot(avg_ArrDelay_Month,
        color='#155ff0',
        alpha = 0.75)

# Setting the title
ax.set_title("Line Plot for Average Arrival Delay by Month",
                color='#2c2f33',
                family = 'monospace',
                size = 25,
                weight = 'bold')

# Creating the x-axis labels
ax.set_xlabel('Months',
                 color='#2c2f33',
                 family = 'monospace',
                 size = 18,
                 labelpad=7)

# Creating the y-axis labels
ax.set_ylabel('Average Arrival Delay (Time in minutes)',
                 color='#2c2f33',
                 family = 'monospace',
                 size = 18,
                 labelpad=10)

# Formatting the x and y axis tick labels
ax.tick_params(axis='x', labelcolor='#ef4343', labelsize=15)
ax.tick_params(axis='y', labelcolor='#ffc000', labelsize=15)
plt.xticks(np.arange(1, 13, 1))

# Adjusting the subplot layout parameters
plt.subplots_adjust(left=0.15)

# Constructing the overall average arrival delay variable
avgDelay = np.sum(full_dataset['ArrDelay']) / len(full_dataset['ArrDelay'])
# Plotting the overall average arrival delay variable so that baseline comparisons can be made easily
plt.axhline(y = avgDelay, color='#f44336', label = 'Overall average delay', linestyle = 'dashed')

# Showing the legend
plt.legend()

# Saving the graph
plt.savefig("Line plot for average arrival delay by month.png")
plt.show()
```

### Constructing a line plot for the causal factors of average delay by month

```python
# Line plot for cause of average delays by month
fig, ax = plt.subplots(1, 1, figsize = (8, 8))

# Filtering the dataset to obtain relevant delays
delay_month_df = full_dataset.filter(['Month',
                                      'CarrierDelay',
                                      'WeatherDelay',
                                      'NASDelay',
                                      'SecurityDelay',
                                      'LateAircraftDelay'], axis=1)

# Making a dataframe which holds the average CarrierDelay, WeatherDelay, NASDelay, SecurityDelay and LateAircraftDelay grouped by month
delay_month_df = delay_month_df.groupby('Month')[['CarrierDelay', 
                                                  'WeatherDelay', 
                                                  'NASDelay', 
                                                  'SecurityDelay', 
                                                  'LateAircraftDelay']].mean()
# Plotting the line graphs
ax.plot(delay_month_df,
        alpha = 0.75)

# Setting the title
ax.set_title("Line Plot for Causal Factors of Average Arrival Delay by Month",
                color='#2c2f33',
                family = 'monospace',
                size = 20,
                weight = 'bold')

# Creating the x-axis labels
ax.set_xlabel('Months',
                 color='#2c2f33',
                 family = 'monospace',
                 size = 18,
                 labelpad=7)

# Creating the y-axis labels
ax.set_ylabel('Average Arrival Delay (Time in minutes)',
                 color='#2c2f33',
                 family = 'monospace',
                 size = 18,
                 labelpad=10)

# Formatting the x and y axis tick labels
ax.tick_params(axis='x', labelcolor='#2c2f33', labelsize=15)
ax.tick_params(axis='y', labelcolor='#2c2f33', labelsize=15)
plt.xticks(np.arange(1, 13, 1))

# Formatting the legend
ax.legend(delay_month_df, loc='upper center', bbox_to_anchor=(0.5, 1.25), ncol=3, fancybox=True, borderpad=1)

# Adjusting the subplot layout parameters
plt.subplots_adjust(left=0.15)

# Saving the graph
plt.savefig("Line plot for cause of average delay by month.png")
plt.show()
```

# Best time of day to fly to minimize delays

Checking for errors in CRSDepTime

We use CRSDepTime since customers can only choose from scheduled departure times

```python
#Checking the range of CRSDepTime
print("The maximum CRSDepTime is", full_dataset.CRSDepTime.max())
print("The minimum CRSDepTime is", full_dataset.CRSDepTime.min())

#Checking the type of CRSDepTime
print("The dtype of CRSDepTime is", full_dataset['CRSDepTime'].dtype)
```

Making 8 bins which are spaced 3 hours apart to categorize each time of day 

```python
# Making a list which holds 8 intervals of different times of day of their corresponding Arrival Delays
delay_TimeOfDay = [full_dataset[(full_dataset.CRSDepTime >= 0000.0) & (full_dataset.CRSDepTime < 0300.0)].ArrDelay, 
                   full_dataset[(full_dataset.CRSDepTime >= 0300.0) & (full_dataset.CRSDepTime < 0600.0)].ArrDelay, 
                   full_dataset[(full_dataset.CRSDepTime >= 0600.0) & (full_dataset.CRSDepTime < 0900.0)].ArrDelay,
                   full_dataset[(full_dataset.CRSDepTime >= 0900.0) & (full_dataset.CRSDepTime < 1200.0)].ArrDelay,
                   full_dataset[(full_dataset.CRSDepTime >= 1200.0) & (full_dataset.CRSDepTime < 1500.0)].ArrDelay,
                   full_dataset[(full_dataset.CRSDepTime >= 1500.0) & (full_dataset.CRSDepTime < 1800.0)].ArrDelay,
                   full_dataset[(full_dataset.CRSDepTime >= 1800.0) & (full_dataset.CRSDepTime < 2100.0)].ArrDelay,
                   full_dataset[(full_dataset.CRSDepTime >= 1800.0) & (full_dataset.CRSDepTime <= 2400.0)].ArrDelay]
```

Constructing a box plot for arrival delay by time of day

```python
# Box plot for arrival delay by time of day
sns.set(style = 'whitegrid')
fig, ax = plt.subplots(figsize = (20,16))

# Plotting 8 vertical boxplots for each time of day 
bp = ax.boxplot(delay_TimeOfDay, vert=True, 
                flierprops=dict(markerfacecolor='r', marker='s', markeredgecolor='white'),
                showmeans=True)

# Creating the y-axis labels
ax.set_ylabel('Arrival Delay (Time in Minutes)',
                 color='#404040',
                 family = 'monospace',
                 size = 18,
                 labelpad=10)

# Creating the x-axis labels
ax.set_xlabel('Time of Day',
                 color='#404040',
                 family = 'monospace',
                 size = 18,
                 labelpad=7)

# Setting the title
ax.set_title("Box-plot for Arrival Delay by Time of Day",
                color='#404040',
                family = 'monospace',
                size = 25,
                weight = 'bold')

# restricting y axis range to clearly view the box-plots
ax.set_ylim([-75,75])

# Assigning x-axis labels
xvalues = ["0000h-0300h", "0300h-0600h", "0600h-0900h", "0900h-1200h", "1200h-1500h", "1500h-1800h", "1800h-2100h", "2100h-0000h"]

# Setting the xvalues variable created above as xtick values
plt.xticks(1 + np.arange(8), xvalues)
plt.yticks()

# Formatting x and y tick labels
ax.tick_params(axis='x', labelcolor='#404040', labelsize=15)
ax.tick_params(axis='y', labelcolor='#ffc000', labelsize=15)

# Adjusting the subplot layout parameters
plt.subplots_adjust(left=0.15)

# Saving the graph
plt.savefig("Boxplot for arrival delay by time of day.png")
plt.show()
```

Obtaining summary statistics for above boxplot 

Median of each box plot

```python
# Iterating over each median value in the bp variable and printing the result
for x in bp['medians']:
        linestat = x.get_ydata()
        median = linestat[0]
        print(median)
```

Mean of each box plot

```python
# Iterating over each mean value in the bp variable and printing the result
for x in bp['means']:
        linestat = x.get_ydata()
        mean = linestat[0]
        print(mean)
```
